### Predict using Supervised ML
###### Predict the percentage of a student based on the no. of study hours

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('first_task.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
print(dataset.info())
```

```python
print(dataset.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
plt.scatter(X_train, y_train, color = 'red')
plt.scatter(X_test, y_test, color = 'red')
plt.title('Score vs No. of hours studied')
plt.xlabel('No. of hours studied')
plt.ylabel('Score')
```

```python
dataset.describe()
```

```python
dataset.corr()
```

```python
import seaborn as sns
sns.set(style="white",palette="BuPu_r")
sns.distplot(dataset['Hours'])
plt.xlabel('Hours')
plt.ylabel('Probability')
plt.title('Distribution Plot')
```

```python
axis = sns.heatmap(dataset.corr(),annot=True)
bottom,top = axis.get_ylim()
axis.set_ylim(bottom,top)
```

```python
sns.set(style="white",palette="gist_heat_r")
sns.pairplot(dataset)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(X_test)
y_pred
```

```python
# Visualising the Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Score vs No. of hours studied (Training set)\n Scatter plot- Actual \nline plot-predicted')
plt.xlabel('No. of hours studied')
plt.ylabel('Score')
```

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, regressor.predict(X_test), color = 'blue')
plt.title('Score vs No. of hours studied (Test set) \n Scatter plot- Actual \nline plot-predicted')
plt.xlabel('No. of hours studied')
plt.ylabel('Score')
```

```python
print("Predicted score if a student studies for 9.25 hrs/ day is : " + str(round(regressor.predict([[9.25]])[0],2)))
```
